```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
sns.set()
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

# Load Data

```python
data = pd.read_csv('1.01. Simple linear regression.csv')
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

# Regression

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.275 + 0.0017*x1
fig = plt.plot(x1,yhat,lw=1,c='orange',label='regression line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
data_dummy = pd.read_csv('1.03. Dummies.csv')
```

```python
data_dummy.head()
```

```python
data_cat = data_dummy.copy()
```

```python
data_cat['Attendance'] = data_cat['Attendance'].map({'Yes':1,'No':0})
```

```python
data_cat.head()
```

```python
data_cat.describe()
```

##### Regression

```python
y = data_cat['GPA']
x1 = data_cat[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data_cat['SAT'],y) #Scatter for data_cat 
y_no = 0.6439 + 0.0014*data_cat['SAT'] 
y_yes = (0.6439+0.2226) + 0.0014*data_cat['SAT']
fig = plt.plot(data_cat['SAT'],y_no,lw=1,c='red')
fig = plt.plot(data_cat['SAT'],y_yes,lw=1,c='green')
plt.xlabel('SAT',fontsize = 20)
plt.xlabel('GPA',fontsize = 20)
plt.show
```

```python
new_data = pd.DataFrame({'const':1,'SAT':[1700,1670],'Attendance':[0,1]})
new_data = new_data[['const','SAT','Attendance']]
new_data
```

```python
predictions = results.predict(new_data)
predictions
```
